# Iris Classification with K-Nearest Neighbors (KNN)

## Goal of this notebook
Use a simple **KNN classifier** to solve a **classification** task on the **Iris** dataset.

We will follow the standard supervised learning workflow:

1. **Understand the task** (features vs. target)  
2. **Load the dataset**  
3. **Train/Test split** (test set = **20%**)  
4. **Train** a KNN model (**k = 5**)  
5. **Predict** on the test set  
6. **Evaluate** using accuracy

![Image description](https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Machine+Learning+R/iris-machinelearning.png)

## 1) What is the Iris dataset?

The Iris dataset is a classic dataset for supervised learning.

### Task
Given measurements of a flower, predict its **species**.

### Features (X)
Each flower has 4 numeric features:

- sepal length (cm)
- sepal width (cm)
- petal length (cm)
- petal width (cm)

### Target (y)
The target is the species (a categorical label with 3 classes):

- setosa
- versicolor
- virginica

So this is a **classification** problem.


## 2) Import libraries

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import pandas as pd
```

## 3) Load and inspect the dataset

```python
iris = load_iris()

X = iris.data       # features
y = iris.target     # target labels

print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Create a DataFrame for a quick look
df = pd.DataFrame(X, columns=iris.feature_names)
df["target"] = y
df.head()
```

## 4) Train/Test split (test = 20%)

We will hold out **20%** of the data as a **test set** (unseen during training).

- Training set: used to fit the model  
- Test set: used only to evaluate performance
- random_state=42:
This fixes the random split so that we get the same train and test sets every time we run the code. We use 42 as the random seed to make the data split reproducible. The specific number doesn’t matter. It just fixes the randomness so we get the same result every time.
- stratify=y:
This makes sure that the class proportions in the training and test sets are similar to those in the original dataset.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Train size:", X_train.shape, " Test size:", X_test.shape)
```

## 5) Train a KNN classifier (k = 5)

Here we choose **k = 5** as a simple example.

- `k` (number of neighbors) is a **hyperparameter**
- We are **not** tuning k in this notebook (that would be the next, more advanced step)


```python
knn = KNeighborsClassifier(n_neighbors=5)  # k = 5
knn.fit(X_train, y_train)
```

## 6) Predict on the test set

```python
y_pred = knn.predict(X_test)
y_pred[:10]
```

## 7) Evaluate the model

```python
test_acc = accuracy_score(y_test, y_pred)
print(f"Test accuracy (k=5): {test_acc:.3f}")
```

```python
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

## 8) Summary

In this notebook, we:

- Introduced the Iris dataset (task, features, target)
- Split data into train/test (test = 20%)
- Trained a KNN classifier with **k = 5**
- Evaluated performance on the test set
